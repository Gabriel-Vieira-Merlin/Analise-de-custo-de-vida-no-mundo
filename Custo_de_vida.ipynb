```python
#Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

#Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

#Lib para ignorar avisos
import warnings

#Lib para transformar dados em números
from sklearn.preprocessing import LabelEncoder

#Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#Realizando a leitura da base de dados
base_dados = pd.read_csv('cost_of_living.csv')
```

```python
#Gráfico em barras comparando o custo de vida de cada país

plt.figure(figsize=(20,10)) #Alterando o tamanho da figura
plt.xlabel('País') #Mudando o nome do eixo x
plt.ylabel('Custo de vida') #Mudando o nome do eixo y
plt.xticks(rotation= 90) #Definindo a rotação dos textos do eixo x
plt.bar(base_dados['country'], base_dados['cost_index'], color= ['red', 'blue']); #Plotando o gráfico de barras
```

```python
#Gráfico em barras comparando a renda mensal de cada país

plt.figure(figsize=(20,10)) #Alterando o tamanho da figura
plt.xlabel('País') #Mudando o nome do eixo x
plt.ylabel('Renda mensal') #Mudando o nome do eixo y
plt.xticks(rotation= 90) #Definindo a rotação dos textos do eixo x
plt.bar(base_dados['country'], base_dados['monthly_income'], color= ['green', 'Yellow']); #Plotando o gráfico de barras
```

```python
#Gráfico em barras comparando o poder de compra de cada país

plt.figure(figsize=(20,10)) #Alterando o tamanho da figura
plt.xlabel('País') #Mudando o nome do eixo x
plt.ylabel('Poder de compra') #Mudando o nome do eixo y
plt.xticks(rotation= 90) #Definindo a rotação dos textos do eixo x
plt.bar(base_dados['country'], base_dados['purchasing_power_index'], color= ['purple', 'orange']); #Plotando o gráfico de barras
```

```python
#Gráfico scatter relacionando a renda mensal com o poder de compra de todos os países

#Armazenando os 27 primeiros países para melhor visualização do gráfico
base_dados_1 = base_dados.sort_values(by= 'monthly_income')[:27]
base_dados_2 = base_dados.sort_values(by= 'monthly_income')[27:54]
base_dados_3 = base_dados.sort_values(by= 'monthly_income')[54:81]
base_dados_4 = base_dados.sort_values(by= 'monthly_income')[81:]

#Paleta de cores (cada ponto terá uma cor diferente)
paleta = sns.color_palette("husl", n_colors=27)
paleta_2 = sns.color_palette("husl", n_colors=26)

#Criando os subplots
fig, ax = plt.subplots(4, figsize=(20, 20)) #Criação de 4 subplots disposto em 2 linhas e 2 colunas com cada um tendo 15 de largura e 10 de altura

#Adicionando título
fig.suptitle('Análise de Renda Mensal e Índice de Poder de Compra')

#Primeiro subplot
ax[0].scatter(base_dados_1['monthly_income'], base_dados_1['purchasing_power_index'], color=paleta)
#Adicionando texto ao final de cada marcador
for index, row in base_dados_1.iterrows(): #Percorre cada linha do dataframe, é necessário usar o iterrows para iterar as linhas e não as colunas, coloca as variáveis index e row, index representa o índice (1, 2, 3...) e row é o valor de cada índice
   ax[0].text(
       row['monthly_income'], #Coordenada x
       row['purchasing_power_index'], #Coordenada y
       row['country'] #Define qual texto será mostrado
   )
#Adicionando nomes para o eixo x e y
ax[0].set_xlabel('Renda mensal') #Nome do eixo x
ax[0].set_ylabel('Poder de compra') #Nome do eixo y

#Segundo subplot
ax[1].scatter(base_dados_2['monthly_income'], base_dados_2['purchasing_power_index'], color=paleta)
#Adicionando texto ao final de cada marcador
for index, row in base_dados_2.iterrows(): #Percorre cada linha do dataframe, é necessário usar o iterrows para iterar as linhas e não as colunas, coloca as variáveis index e row, index representa o índice (1, 2, 3...) e row é o valor de cada índice
   ax[1].text(
       row['monthly_income'], #Coordenada x
       row['purchasing_power_index'], #Coordenada y
       row['country'] #Define qual texto será mostrado
   )
#Adicionando nomes para o eixo x e y
ax[1].set_xlabel('Renda mensal') #Nome do eixo x
ax[1].set_ylabel('Poder de compra') #Nome do eixo y

#Terceiro subplot
ax[2].scatter(base_dados_3['monthly_income'], base_dados_3['purchasing_power_index'], color=paleta)
#Adicionando texto ao final de cada marcador
for index, row in base_dados_3.iterrows(): #Percorre cada linha do dataframe, é necessário usar o iterrows para iterar as linhas e não as colunas, coloca as variáveis index e row, index representa o índice (1, 2, 3...) e row é o valor de cada índice
   ax[2].text(
       row['monthly_income'], #Coordenada x
       row['purchasing_power_index'], #Coordenada y
       row['country'] #Define qual texto será mostrado
   )
#Adicionando nomes para o eixo x e y
ax[2].set_xlabel('Renda mensal') #Nome do eixo x
ax[2].set_ylabel('Poder de compra') #Nome do eixo y

#Quarto subplot
ax[3].scatter(base_dados_4['monthly_income'], base_dados_4['purchasing_power_index'], color=paleta_2)
#Adicionando texto ao final de cada marcador
for index, row in base_dados_4.iterrows(): #Percorre cada linha do dataframe, é necessário usar o iterrows para iterar as linhas e não as colunas, coloca as variáveis index e row, index representa o índice (1, 2, 3...) e row é o valor de cada índice
   ax[3].text(
       row['monthly_income'], #Coordenada x
       row['purchasing_power_index'], #Coordenada y
       row['country'] #Define qual texto será mostrado
   )
#Adicionando nomes para o eixo x e y
ax[3].set_xlabel('Renda mensal') #Nome do eixo x
ax[3].set_ylabel('Poder de compra'); #Nome do eixo y
```
